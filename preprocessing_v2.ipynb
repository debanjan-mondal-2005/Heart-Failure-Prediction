```python
# fill the missing values
data['RestingBP'].fillna(data['RestingBP'].median(), inplace=True)
data['Cholesterol'].fillna(data['Cholesterol'].median(), inplace=True)
```

```python
def cap_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return np.clip(col, lower, upper)

data['Cholesterol'] = cap_outliers(data['Cholesterol'])
data['RestingBP'] = cap_outliers(data['RestingBP'])
```

```python
num_cols = ['Age', 'MaxHR', 'Oldpeak']

def cap_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return np.clip(col, lower, upper)

for col in num_cols:
    data[col] = cap_outliers(data[col])
```
